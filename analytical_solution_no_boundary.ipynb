```python
# importing numpy
import numpy as np

# line used to avoid giving an warning when division by zero?
# this is to supress the warning when division by zero
# division by zero happens for the first time coordinate of time = 0
# np.seterr(divide='ignore', invalid='ignore')

# importing matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

# style used for plotting figures
plt.style.use('styles/mainstyle.use')
```

```python
# Defining Parameters

D = 10**(-8)   # cm2/s - Diffusion Coefficient
v = 0.         # cm/s  - electrophoretic velocity

pi = np.pi     # defining pi
```

```python
# defining space domain and discretization

xi = -0.1                 # initial x value
xf = 0.1                  # final x value
dx = 0.0001               # space step

nx = int((xf-xi)/dx)      # no. of space divisions

x = np.empty(nx+1)        # initializing the space array (initial and final x value included)

for k in range(nx+1):     # building the sapce array
    x[k] = xi + k*dx

print('Inital x coordinate:', x[0], 'cm')                # initial space coordinate
print('Final x coordinate:', x[-1], 'cm')                # final space coordinate
print('No. of elements in space array:', len(x))         # length of the space array (inital and final space included)
```

```python
# defining time domain and discretization

ti = 0.                   # initial time
tf = 100.                 # final time at which solution is needed
dt = 1                    # time step

nt = int((tf-ti)/dt)      # no. of time divisions

t = np.empty(nt+1)        # initialising the time array (initial and final time included)

for k in range(nt+1):     # building the time array
    t[k] = ti + k*dt

print('Initial time:',t[0], 's')                        # initial time coordinate
print('Final time:',t[-1], 's')                         # final time in time coordinate
print('No. of elements in time array:', len(t))         # length of the time array (initial and final time included)
```

```python
# defining probability

# initializing the the p(x,t) array 
# no. of rows as no. of space coordinates
# no. of cols as no. of time coorinates
pxt = np.empty((nx+1,nt+1))    

# initial condition - narrow band gaussian distribution
mu=0.
sigma=0.000125
pxt[:,0] = (1/(np.sqrt(2*np.pi*(sigma**2))))*(np.exp((-1/2)*((x-mu)/(sigma))**2))

for k in range(1,nt+1):
    pxt[:,k]= (1/np.sqrt(4*pi*D*t[k]))*(np.exp(((-(x-v*t[k])**2)/(4*D*t[k]))))

print('No. of rows in our p(x,t) array:', len(pxt[:,1]))
print('No. of cols in our p(x,t) array:', len(pxt[1,:]))
```

```python
fig, ax = plt.subplots(figsize=(8,6))

ax.plot(x,pxt[:,0],'-',label='0s')
ax.plot(x,pxt[:,10],'-k',label='10s')
ax.plot(x,pxt[:,30],'--k',label='30s')
ax.plot(x,pxt[:,60],'-.k',label='60s')

#ax.grid()

ax.set_xlim(-0.003,0.003)
ax.set_ylim(0,3500)
ax.legend(loc='upper left')

ax.set_xlabel('x (cm)')
ax.set_ylabel('P(x,t)')
ax.set_title('Fokker-Planck solution for no drift')
```
